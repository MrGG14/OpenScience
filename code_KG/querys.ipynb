```python
from rdflib import Graph
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
def quitar_espacios(cadena):
    return cadena.replace(' ', '_')
```

## Cargamos el grafo

```python
g = Graph()

# Cargar el archivo RDF (reemplaza 'yourfile.rdf' con el nombre de tu archivo)
g.parse("knowledge_graph_linked.rdf", format="xml")
```

## Realizamos las Querys

### Obtener el t√≠tulo de todos los Papers

```python
query = """
PREFIX sc: <http://schema.org/>

SELECT ?title
WHERE {
  ?paper a sc:Paper .
  ?paper sc:Title ?title .

}
"""
results = g.query(query)
for row in results:
    print(row.title)
```

### Obtener todas las relaciones de un paper en concreto

```python
query_title = "DINOv2 Learning Robust Visual Featu"
query = f"""
PREFIX ex: <http://example.org/>

SELECT ?subject ?predicate ?object
WHERE {{
  ?subject ?predicate ?object .
  FILTER(?subject = ex:{quitar_espacios(query_title)})
}}
"""
results = g.query(query)
for row in results:
    print(f"({row.subject}, {row.predicate}, {row.object})")
```

### Obtener el abstract de un paper en concreto

```python
query_title = "DINOv2 Learning Robust Visual Featu"
query = f"""
PREFIX ex: <http://example.org/>
PREFIX sc: <http://schema.org/>

SELECT ?abstract
WHERE {{
  ex:{quitar_espacios(query_title)} sc:abstract ?abstract.
}}
"""
results = g.query(query)
for row in results:
    print(row.abstract)
```

### Obtener los autores de un paper en concreto

```python
query_title = "tft"
query = f"""
PREFIX ex: <http://example.org/>
PREFIX sc: <http://schema.org/>

SELECT ?author_name
WHERE {{
  ex:{quitar_espacios(query_title)} sc:author ?author.
  ?author sc:name ?author_name
}}
"""
results = g.query(query)
for row in results:
    print(row.author_name)
```


### Obtener las organizaciones referenciadas de un paper en concreto

### Obtener todos los papers que tienen un topic en concreto

```python
query_topic = "deep visual methods local network"
query = f"""
PREFIX ex: <http://example.org/>
PREFIX sc: <http://schema.org/>

SELECT ?title
WHERE {{
  ?paper sc:hasTopic ex:{quitar_espacios(query_topic)}.
  ?paper sc:Title ?title .
}}
"""
results = g.query(query)
for row in results:
    print(row.title)
```

### Obtener los papers similares a uno en concreto

```python
query_paper = "DINOv2 Learning Robust Visual Featu"
query = f"""
PREFIX ex: <http://example.org/>
PREFIX sc: <http://schema.org/>

SELECT ?title
WHERE {{
  ?paper sc:isSimilar ex:{quitar_espacios(query_paper)}.
  ?paper sc:Title ?title .
}}
"""
results = g.query(query)
for row in results:
    print(row.title)
```

### Obtener autores que adem√°s est√°n referenciados en otros papers

```python
query = """
PREFIX sc: <http://schema.org/>

SELECT ?author_name ?paper_titulo ?otro_paper_titulo
WHERE {
  ?paper sc:author ?author .
  ?paper sc:Title ?paper_titulo .
  ?otro_paper sc:acknowledgesPerson ?author .
  ?otro_paper sc:Title ?otro_paper_titulo .
  ?author sc:name ?author_name .

}
"""
results = g.query(query)
for row in results:
    print(f'{row.author_name}, autor de "{row.paper_titulo}", aparece referenciado en "{row.otro_paper_titulo}"end=')
```

### Obtener donde se formaron los autores de un paper en concreto (usando Wikidata)

```python
query_title = "Learning Transferable Visual Models From Natural Language Supervis"

def buscar_pais_de_autor_wikidata(id_autor):
  query = f"""
  SELECT ?country ?countryLabel
  WHERE {{
    wd:{id_autor} wdt:P69 ?country .
    SERVICE wikibase:label {{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }}
  }}
  """
  sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
  sparql.setQuery(query)
  sparql.setReturnFormat(JSON)
  results = sparql.query().convert()
  for result in results["results"]["bindings"]:
      country_label = result["countryLabel"]["value"]
      return country_label


query = f"""
PREFIX ex: <http://example.org/>
PREFIX sc: <http://schema.org/>
PREFIX owl: <http://www.w3.org/2002/07/owl#>

SELECT ?author_name ?wikidata_uri
WHERE {{
  ex:{quitar_espacios(query_title)} sc:author ?author.
  ?author sc:name ?author_name.
  ?author owl:sameAs ?wikidata_uri.
}}
"""
results = g.query(query)
for row in results:
    print(row.author_name, buscar_pais_de_autor_wikidata(row.wikidata_uri[30:]))
```

### Obtener los pa√≠ses de los autores de un paper en concreto (usando Wikidata)

```python
query_title = "DINOv2 Learning Robust Visual Featu"

def buscar_pais_de_autor_wikidata(id_autor):
  query = f"""
  SELECT ?country ?countryLabel
  WHERE {{
    wd:{id_autor} wdt:P27 ?country .
    SERVICE wikibase:label {{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }}
  }}
  """
  sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
  sparql.setQuery(query)
  sparql.setReturnFormat(JSON)
  results = sparql.query().convert()
  for result in results["results"]["bindings"]:
      country_label = result["countryLabel"]["value"]
      return country_label


query = f"""
PREFIX ex: <http://example.org/>
PREFIX sc: <http://schema.org/>
PREFIX owl: <http://www.w3.org/2002/07/owl#>

SELECT ?author_name ?wikidata_uri
WHERE {{
  ex:{quitar_espacios(query_title)} sc:author ?author.
  ?author sc:name ?author_name.
  ?author owl:sameAs ?wikidata_uri.
}}
"""
results = g.query(query)
for row in results:
    print(row.author_name, buscar_pais_de_autor_wikidata(row.wikidata_uri[30:]))
```

### Obtener la edad de los autores de un paper en concreto (usando Wikidata)

```python
query_title = "DINOv2 Learning Robust Visual Featu"

def buscar_edad_de_autor_wikidata(id_autor):
    query = f"""
    SELECT ?fecha_nacimiento 
    WHERE {{
    wd:{id_autor} wdt:P569 ?fecha_nacimiento .
    SERVICE wikibase:label {{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }}
    }}
    """
    sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    if len(results["results"]["bindings"]) > 0:
        for result in results["results"]["bindings"]:
            fecha_nacimiento = result["fecha_nacimiento"]["value"]
            return fecha_nacimiento
    else:
        return '2024'


query = f"""
PREFIX ex: <http://example.org/>
PREFIX sc: <http://schema.org/>
PREFIX owl: <http://www.w3.org/2002/07/owl#>

SELECT ?author_name ?wikidata_uri
WHERE {{
  ex:{quitar_espacios(query_title)} sc:author ?author.
  ?author sc:name ?author_name.
  ?author owl:sameAs ?wikidata_uri.
}}
"""
results = g.query(query)
for row in results:
    print(row.author_name, 2024 - int(buscar_edad_de_autor_wikidata(row.wikidata_uri[30:])[:4]))
```
